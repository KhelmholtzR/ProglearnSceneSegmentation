```python
directory_to_extract_to = './'
with zipfile.ZipFile('ade20k-small.zip', 'r') as zip_ref:
    zip_ref.extractall(directory_to_extract_to)
```

```python
datadir = './ade20k-small/'
imgpaths = [datadir+im for im in os.listdir(datadir) if '.jpg' in im]
lblpaths = [datadir+im for im in os.listdir(datadir) if 'seg' in im]
imgpaths.sort()
lblpaths.sort()
imgpaths, lblpaths
```

```python
# load samples
fig, ax = plt.subplots(5,2)
for i in range(5):
    img = cv2.imread(imgpaths[i])
    lbl = cv2.imread(lblpaths[i]) 
    ax[i][0].imshow(img)
    ax[i][0].axis('off')
    ax[i][1].imshow(lbl)
    ax[i][1].axis('off')
plt.show()
```

```python
# load samples
fig, ax = plt.subplots(1,2)
img = cv2.imread(imgpaths[0])
lbl = cv2.imread(lblpaths[0]) 
ax[0].imshow(img)
ax[0].axis('off')
ax[1].imshow(lbl)
ax[1].axis('off')
plt.show()
```
